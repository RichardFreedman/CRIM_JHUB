```python
# A function needed below

def convertTuple(soggetti):
    tup = soggetti[0]
    str = '_'.join(tup)
    return str

def heatmap(piece):
    nr = piece.notes(combineUnisons=True)
    mel = piece.melodic(df=nr, kind='d', end=False)
    mel_ngrams = piece.ngrams(df=mel, n=3)
    entry_ngrams = piece.entries(df=mel, n=3)
    mel_ngrams_duration = piece.durations(df=mel, n=3, mask_df=entry_ngrams)
    return viz.plot_ngrams_heatmap(entry_ngrams, mel_ngrams_duration, selected_patterns=[], voices=[])
```

### Now Check the Soggetti and Modules.  Any Repetitions across the corpus?

```python
corpus_list = []
for name in glob.glob('Music_Files/*'):
    corpus_list.append(name)
corpus = CorpusBase(corpus_list)
```

```python
# Here are Melodic Ngrams for Corpus and show which ones occur multiple times across the corpus

func = ImportedPiece.melodic

list_of_mels = corpus.batch(func=func, kwargs={'kind': 'd', 'end': False}, metadata=False)
func2 = ImportedPiece.ngrams
list_of_mel_ngrams = corpus.batch(func=func2, kwargs={'df': list_of_mels, 'n': 5, 'offsets': 'last'}, metadata=True)

func3 = ImportedPiece.detailIndex
list_of_details = corpus.batch(func=func3, kwargs={'offset': True, 'df': list_of_mel_ngrams})

mel_corpus = pd.concat(list_of_details)#.fillna('-').dropna(how="all")

c = mel_corpus['Composer']
t = mel_corpus["Title"]
mel_corpus.pop("Composer")
mel_corpus.pop("Title")
mel_corpus.insert(0, "Composer", c)
mel_corpus.insert(1, "Title", t)
mc = mel_corpus.set_index(['Composer', 'Title'])

stacked = mc.stack().dropna().to_frame()
groupstack = stacked.groupby([0, 'Title']).size().reset_index(name='counts')
groupstack.rename(columns={0: 'Module'}, inplace=True)
popular_mods = groupstack[groupstack['counts'] > 30]
pop_list = popular_mods["Module"].to_list()
for popular_mod in pop_list:
    print(groupstack[groupstack['Module'] == popular_mod])
    
# groupstack[groupstack['Module'] == ('4', '1', '2', '2', '-3')]
```

```python
# Search for the melodies listed above:


def convertTuple(tup):
    out = ""
    if isinstance(tup, tuple):
        out = ', '.join(tup)
    return out

@interact
def mel_ngram_search(my_search="", df = fixed(mel_corpus)):
    df_no_tuple = df.applymap(convertTuple)
    df_no_tuple.pop("Composer")
    df_no_tuple.pop("Title")
    df_no_tuple.insert(0, "Composer", df["Composer"])
    df_no_tuple.insert(1, "Title", df["Title"])
    filtered_ngrams = df_no_tuple[df_no_tuple.apply(lambda x: x.astype(str).str.contains(my_search).any(), axis=1)].copy()
    
    pd.set_option('max_columns', None)
    return filtered_ngrams.fillna("-").reset_index().applymap(str).style.applymap(lambda x: "background: #ccebc4" if re.search(my_search, x) else "")
```

#### Count Modules across the corpus

```python
# This is the corpus as modular ngrams showing those that occur more than twice
# could be any length of ngram!

func = ImportedPiece.ngrams
list_of_modules = corpus.batch(func=func, kwargs={'n': 5, 'interval_settings': ('d', True, True), 'offsets': 'last'}, metadata=True)
func2 = ImportedPiece.detailIndex
list_of_details = corpus.batch(func=func2, kwargs={'offset': True, 'df': list_of_modules})

module_corpus = pd.concat(list_of_details)#.fillna('-').dropna(how="all")

c = module_corpus['Composer']
t = module_corpus["Title"]
module_corpus.pop("Composer")
module_corpus.pop("Title")
module_corpus.insert(0, "Composer", c)
module_corpus.insert(1, "Title", t)
mc2 = module_corpus.set_index(['Composer', 'Title']).copy()
stacked = mc2.stack().dropna().to_frame()
modules_grouped_stacked = stacked.groupby([0, 'Title']).size().reset_index(name='counts')
modules_grouped_stacked.rename(columns={0: 'Module'}, inplace=True)
modules_grouped_stacked[modules_grouped_stacked['counts'] > 2]
```

```python
# Here you can search for any of the modules listed above, and find where they are in each piece!
@interact
def new_ngram_search(my_search="", df = fixed(module_corpus)):
    df2 = module_corpus.copy()
    c = df['Composer']
    t = df["Title"]
    df2.pop("Composer")
    df2.pop("Title")
    df2.insert(0, "Composer", c)
    df2.insert(1, "Title", t)
    filtered_ngrams = df2[df2.apply(lambda x: x.astype(str).str.contains(my_search).any(), axis=1)].copy()
    pd.set_option('max_columns', None)
    output =  filtered_ngrams.fillna("-").reset_index().applymap(str).style.applymap(lambda x: "background: #ccebc4" if re.search(my_search, x) else "")
    return output
```
